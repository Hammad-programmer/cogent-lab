##Prompt Template

```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate.from_template("""Create a personalized study plan for {Student}, who is a {Year_of_Study} student in {Field_of_Study}.

**Student Information:**
- Subjects: {Subjects}
- Learning Styles: {Learning_Styles}
- Personal Objectives: {Personal_Objectives}
- Challenges: {Challenges}
- Extracurricular Activities: {Extracurricular_Activities}

**Plan Requirements:**
- Consider student's strengths and weaknesses in each subject.
- Allocate time for each subject based on difficulty and workload.
- Suggest specific study activities aligned with learning styles.
- Incorporate recommended resources beyond textbooks (e.g., videos, quizzes, simulations).
- Offer strategies to address challenges and learning difficulties.
- Integrate extracurricular activities without compromising study time.
- Design goal-oriented activities to achieve personal objectives.
- Include progress tracking mechanisms for monitoring and adjustment.
- Encourage regular review and evaluation of the plan.

**Prompt:**
Based on the provided student information, please generate a comprehensive and personalized study plan that addresses all of the plan requirements."""
)
```

```python
# Get Hammad's row
hammad_data = df.iloc[0]

prompt = prompt.format(
    Student=hammad_data['Name'],
    Year_of_Study=hammad_data['Year of Study'],
    Field_of_Study=hammad_data['Field of Study'],
    Subjects=hammad_data['List of Subjects'],
    Learning_Styles=hammad_data['Preferred Learning Styles'],
    Personal_Objectives=hammad_data['Personal Objectives'],
    Challenges=hammad_data['Challenges'],
    Extracurricular_Activities=hammad_data['Extracurricular Activities'])
```

```python
example_prompt = PromptTemplate(
    input_variables=["question", "answer"], template="Question: {question}\n{answer}"
)

print(example_prompt.format(**examples[0]))
```

```python
prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    suffix="Question: {input}",
    input_variables=["input"],
)
```
